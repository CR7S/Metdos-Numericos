```python
def funcion(c) :
    g = 9.8
    m = 68.1
    t = 10.0
    v = 40.0
    e = 2.718
    
    return (g*m/c*(1-e**-(c/m*t))-v)
    
xl = 12.0
xu = 20.0
xr = 0.0
ter = -1
et = 0.0

print ("MÉTODO DE BISECCIÓN")
print '|{0:10} |  {1:20} | {2:20} | {3:20} | {4:20} | {5:20} |'.format('Término','Xu','Xr','Xl','F(c)','εt')
while True:
    ter += 1
    if (xr!=0):
        et = (14.7802-xr)/xr*100 
        print '|{0:10} |  {1:20} | {2:20} | {3:20} | {4:20} | {5:15} |'.format(ter, xu, xr, xl, funcion(xr), et)
    else :
        print '|{0:10} |  {1:20} | {2:20} | {3:20} | {4:20} |'.format(ter, xu, xr, xl, et)
    if (funcion(xl)*funcion(xu)<0):
        xr = (xl+xu)/2
        if (funcion(xl)*funcion(xr)<0) :
            xu = xr
            continue
        if (funcion(xl)*funcion(xr)>0):
            xl = xr
            continue
        if (funcion(xl)*funcion(xr)==0):
            print ("LA RAIZ ESTÁ EN t = %1.3f" %(xr))
            break
    else :
        break

print ()
print ("MÉTODO DE LA FALSA POSICIÓN")



xl, xu, xr, ter, et2 =  12.0, 20.0, 0.0, -1, 0.0


print '|{0:10} |  {1:20} | {2:20} | {3:20} | {4:20} | {5:15} |'.format('Término','Xu','Xr','Xl','F(c)','εt')
while True:
    ter += 1
    if (xr!=0):
        et2 = (14.7802-xr)/xr*100
        print '|{0:10} |  {1:20} | {2:20} | {3:20} | {4:20} | {5:15} |'.format(ter, xu, xr, xl, funcion(xr), et)
    else :
        print '|{0:10} |  {1:20} | {2:20} | {3:20} | {4:20} |'.format(ter, xu, xr, xl, et)
    if (funcion(xl)*funcion(xu)<0):
        xr = xu-((funcion(xu)*(xl-xu))/(funcion(xl)-funcion(xu)))
        if (funcion(xl)*funcion(xr)<0):
            xu = xr
            continue
        if (funcion(xl)*funcion(xr)>0):
            xl = xr
            continue
        if (funcion(xl)*funcion(xr)==0):
            print ("LA RAIZ ESTÁ EN t = %1.3f" %(xr))
            break
    else:
        break
        
print ()
print ("-El error relativo verdadero del metodo de Bisección: %1.3f" %(et))
print ("-El error relativo verdadero del metodo de Falsa Posición: %1.3f" %(et2))
```
